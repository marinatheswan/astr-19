```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize the Generator and pull 1000 random integers, in this case it's between 0 and 1000
rng = np.random.default_rng()
data = rng.integers(low=0, high=1000, size=1000)

# Create plot
plt.figure(figsize=(10, 6))
plt.hist(data, bins=100, color='skyblue', edgecolor='black')


plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Integers (100 Bins)')

plt.savefig('Marina_random_histogram.pdf')

plt.show()
```
