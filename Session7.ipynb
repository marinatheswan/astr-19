```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create numpy array ranging from 0 to 1 in 100 steps
x = np.linspace(0, 1, 100)

# function returns exp(x)
def calculate_awesomeness(val):
    return np.exp(val)
```

```python
# Use the function to set y
y = calculate_awesomeness(x)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Growth Curve', color='blue')

# Add labels
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.title('Exponential Growth of Awesomeness')

# Save as PDF
plt.savefig('awesomeness_plot.pdf')

plt.show()
```
