```python
import numpy as np

def get_sin(x):
    return np.sin(x)
```

^This returns sin(x) using numpy^

```python
def get_cos(x):
    return np.cos(x)
```

^This returns cos(x) using numpy^

```python
x_values = np.linspace(0, 2 * np.pi, 1000)

sin_values = get_sin(x_values)
cos_values = get_cos(x_values)
```

^This creates 1000 entries between 0 and 2pi, evenly spaced. It takes the x values from sin and cos.^

```python
# Print the header for the table
print(f"{'x':<10} | {'sin(x)':<10} | {'cos(x)':<10}")
print("-" * 35)

# Use a for loop to print the first 10 values in columns
for i in range(10):
    print(f"{x_values[i]:<10.4f} | {sin_values[i]:<10.4f} | {cos_values[i]:<10.4f}")
```

^It prints headers for the table, and a divider between the values. Then it uses a for loop to print the first 10 tabulated values for sin and cos in columns.^
