```python
import matplotlib.pyplot as plt
import numpy as np

# array from 0 to 1, with 100 evenly spaced numbers
x = np.linspace(0, 1, 100)
```

```python
def get_sin(x):
    return np.sin(x)

def get_cos(x):
    return np.cos(x)
```

```python
# Create a figure with 1 row and 2 columns
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Left panel sin(x)
ax[0].plot(x, get_sin(x), color='blue')
ax[0].set_title('sin(x)')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

# Right panel cos(x)
ax[1].plot(x, get_cos(x), color='red')
ax[1].set_title('cos(x)')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')

# Adjust layout to prevent overlap
plt.tight_layout()

# Save the figure as a PDF
plt.savefig('Marina_trig_plots.pdf')

plt.show()
```
